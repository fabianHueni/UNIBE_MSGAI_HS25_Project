This notebook is to analyse the dataset and to find the ID's of the dataset samples where the quotation marks are missing. this is important for the csv context parsing to work correctly

```python
import pandas as pd
from pathlib import Path

# Path to your CSV file
csv_path = Path("./dataset/boolq_validation.csv")  # change this

# Read raw text (not via pandas' CSV parser yet)
text = csv_path.read_text(encoding="utf-8")

lines = text.replace("\r\n", "\n").replace("\r", "\n").split("\n")
header = lines[0]
data_lines = lines[1:]

bad_ids = []

for line in data_lines:
    if not line.strip():
        continue

    # Split out id, question, answer; rest is passage
    parts = line.split(",", 3)
    if len(parts) < 4:
        # malformed line; can't even get 4 fields
        bad_ids.append(parts[0] if parts else None)
        continue

    pid, q, ans, passage = parts
    trimmed = passage.strip()

    # context/passage should be fully wrapped in double quotes
    if not (trimmed.startswith('"') and trimmed.endswith('"')):
        bad_ids.append(pid)



# print all the id's of the samples with missing quotation marks
bad_ids
```
